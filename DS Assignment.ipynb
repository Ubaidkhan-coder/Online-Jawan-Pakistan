```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
print(data)
```

```python
hp_mean = data["HP"].mean()

def check_hp(val):
    if val < hp_mean:
        return "Low_HP"
    else:
        return "High_HP"
        
```

```python
data["High_Low_HP"] = data["HP"].apply(check_hp)
data
```

```python
speed_mean = data["Speed"].mean()

def check_speed(val):
    if val < speed_mean:
        return "Low Speed"
    else:
        return "High Speed"
```

```python
data["High_Low_Speed"] = data["Speed"].apply(check_speed)
data
```
